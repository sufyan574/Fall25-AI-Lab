<div class="alert alert-block alert-warning">
    4.1. <b>Manage Null Values</b>

```python

df = df.drop('airline', axis=1)
```

```python

df = df.drop('flight', axis=1)
```

```python
price_mean = df['price'].mean()
price_median = df['price'].median()
price_mode = df['price'].mode()[0]

print("\nPrice Mean :", price_mean)
print("Price Median :", price_median)
print("Price Mode :", price_mode)
```

```python
df['source_city'] = df['source_city'].fillna(df['source_city'].mode()[0])
```

```python
df['destination_city'] = df['destination_city'].fillna(df['destination_city'].mode()[0])
```

```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].mean())
```

```python
for col in df.select_dtypes(include=['int64', 'float64']).columns:
    df[col] = df[col].fillna(df[col].mean())
```

```python
df.info()
```

<div class="alert alert-block alert-warning">
    4.2. <b>Manage Data Types</b>
</div>

```python
df['price'] = df['price'].astype('int64')
```

```python
categorical_cols = [
    'source_city',
    'departure_time',
    'stops',
    'arrival_time',
    'destination_city',
    'class'
]

for col in categorical_cols:
    df[col] = df[col].astype(str)
```

```python
print(df.columns.tolist())
```

```python
df['duration'] = df['duration'].astype(float)
df['days_left'] = df['days_left'].astype(int)
df['price']     = df['price'].astype(int)

print(df.dtypes)
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
def convertObjToInt(cols):
    for i in cols:
        data = pd.DataFrame({i:df[i].unique()})
        data_LE = LabelEncoder()
        data_LE.fit(np.ravel(data))    
        df[i] = data_LE.transform(df[i])
        
columns = ['source_city', 'departure_time', 'stops']
convertObjToInt(columns)
```
